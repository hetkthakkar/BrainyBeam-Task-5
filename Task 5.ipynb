## Task-5: Multiple Y-Axes Sharing a Common X-Axis Plot
### Description: Create a plot with multiple y-axes sharing a common x-axis.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for the plot
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create a figure and axis
fig, ax1 = plt.subplots()

# Set background color
fig.patch.set_facecolor('linen')

# Plotting on the first y-axis
ax1.plot(x, y1, color='red', label='sin(x)')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('sin(x)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis
ax2 = ax1.twinx()
ax2.plot(x, y2, color='green', label='cos(x)')
ax2.set_ylabel('cos(x)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Create a third y-axis
ax3 = ax1.twinx()

# Adjust the position of the third y-axis
ax3.spines['right'].set_position(('outward', 60))
ax3.plot(x, y3, color='blue', label='tan(x)')
ax3.set_ylabel('tan(x)', color='blue')
ax3.tick_params(axis='y', labelcolor='blue')

# Title and grid
plt.title('Multiple Y-Axes Sharing a Common X-Axis')
plt.grid(True)

# legend
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9), fontsize=10)

# Display the plot
plt.tight_layout()

# Save the plot as a PNG file
plt.savefig('multiple_y_axes_plot.png')

# Show the plot
plt.show()
```
